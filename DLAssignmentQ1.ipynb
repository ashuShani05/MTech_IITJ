```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Data = {'x1':[1,-1,0,0.1,0.2,0.9],'x2':[1,-1,0.5,0.5,0.2,0.5],'class':[1,-1,-1,-1,1,1]}
wt = np.array([1,1])
x = np.array([Data['x1'],Data['x2']]).T
y = np.array([Data['class']]).T
step = 0
```

```python
# Initial decision boundary
plt.scatter(np.dot(x,wt),y)
plt.xlabel('data points')
plt.ylabel('classes')
plt.show
```

```python
def convergence(x,y,wt,step):
  for i in range(6) :
    if (np.dot(wt.T,x[i])>0 and y[i]>0) or (np.dot(wt.T,x[i])<0 and y[i]<0):
      pass
    else:
      # print(x[i])
      if y[i]>0:
        wt = wt+x[i]
        step+=1
        print('updated wt',wt)
        convergence(x,y,wt,step)
      else:
        # print(x[i])
        wt = wt-x[i]
        step+=1
        print('updated wt',wt)
        convergence(x,y,wt,step)
  return(step,wt)  
```

```python
new_wt = convergence(x,y,wt,step)[1]
step = convergence(x,y,wt,step)[0]
print('\nFinal Decison boundary would be - ',np.dot(x,new_wt))
```

```python
print('\nsteps to converge - {} and updated wt - {}'.format(step,new_wt))
```

```python
# updated decision boundary
plt.scatter(np.dot(x,new_wt),y)
plt.xlabel('data points')
plt.ylabel('classes')
plt.show
```
