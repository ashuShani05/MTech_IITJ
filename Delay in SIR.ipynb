```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
N = 1000
b = 0.3
a = 0.1
w = 10
t_max = 100
```

```python
# Initial conditions
S0 = N - 1
I0 = 1
R0 = 0
```

```python
# Time vector
t = np.linspace(0, t_max, t_max + 1)
```

```python
# Solution arrays
S = np.zeros(t.shape)
I = np.zeros(t.shape)
R = np.zeros(t.shape)
```

```python
# Initial values
S[0] = S0
I[0] = I0
R[0] = R0

# loop
for i in range(1, t.shape[0]):
    S[i] = S[i-1] - b*S[i-1]*I[i-1] + a*I[max(i-w,0)]
    I[i] = I[i-1] + b*S[i-1]*I[i-1] - a*I[i-1]
    R[i] = R[i-1] + a*I[max(i-w,0)] - a*I[i-w]
```

```python
# Plotting
plt.plot(t, S/N, label='Susceptible')
plt.plot(t, I/N, label='Infectious')
plt.plot(t, R/N, label='Recovered')
plt.xlabel('Time (num of days)')
plt.ylabel('Proportion of population')
plt.legend()
plt.show()
```
