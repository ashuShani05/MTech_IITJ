```python
# Importing libraries

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Define the SIR model function
def sir_model(y, t, N, b, g):
    S, I, R = y
    dSdt = -b * S * I / N
    dIdt = b * S * I / N - g * I
    dRdt = g * I
    return dSdt, dIdt, dRdt
```

```python
# Set the initial conditions and parameters
N = 1000 # Total population
I0, R0 = 1, 0 # Initial number of infectious and recovered individuals
S0 = N - I0 - R0 # Initial number of susceptible individuals
b = 0.2 # Effective contact rate
g = 0.1 # Mean recovery rate
```

```python
# Set the time grid
t = np.linspace(0, 100, 1500)
```

```python
# Solve the SIR model using odeint
solution = odeint(sir_model, [S0, I0, R0], t, args=(N, b, g))
```

```python
# Plot the solution curves
plt.plot(t, solution[:, 0], label='Susceptible')
plt.plot(t, solution[:, 1], label='Infectious')
plt.plot(t, solution[:, 2], label='Recovered')
plt.xlabel('Time (num of days)')
plt.ylabel('num of individuals')
plt.title('SIR Model Solution')
plt.legend()
plt.show()
```
